Marbles Game

In this exercise, you will be using the random and matplotlib libraries in Python to write a simulation of the Marbles Game. The rules of the game are as follows: 
There is a bag with 3 marble balls: 1 blue, 1 green, and 1 red. A blue marble is equal to 1 point, a green marble is 3 points, and a red marble is 5 points.
There are at least 2 players.
On each turn, a player draws two marbles from the bag, with replacement. 
If a single green marble is drawn, the player scores nothing, and the turn ends.
If two blue marbles are drawn, the player’s entire score is reset, and the turn ends.
Otherwise, the sum of the marbles is added to the player’s total score, and the turn ends.
Whoever accumulates 23 points in the fewest amount of turns wins.
If more than 1 player accumulates 23 points or more in the same amount of turns, then the game ends in a tie, and all players with 23 points or more win.
Write a function called marbles_game(players) that inputs the number of players (the maximum number of players is 4) and simulates the game. You can accept that the players are not playing simultaneously, but are playing one after the other. This means that Player 1 plays until they reach 23 points, then Player 2 plays until they reach 23 points, etc. 
Your program should produce a line graph that shows the players’ running scores after each turn. The plot should look similar to the one below, but might have a different pattern      
           
Input Format: players: int - the number of players
Output Format: str of a text signaling who won the match
   + graph generated by Matplotlib
Libraries allowed: matplotlib, random

```python
# imports 
import random 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sample():
    marbles = ['blue', 'red', 'green']
    return random.choice(marbles)
```

```python
def simulate():
    # simulate game of marbles
    # single green => nothing and turn ends
    # 2 blue => score reset and turn ends
    # otherwise sum
    score = [0]
    n = 0
    values = {'blue': 1, 'green': 3, 'red': 5}
    
    while score[n] < 23:
        first = sample()
        second = sample()
        if first == 'green' or second == 'green' and second != first:
            score.append(score[n])
        elif second == first and second == 'blue':
            score.append(0)
        else:
            score.append(score[n] + values[first] + values[second])
        n += 1
    return score
        
        
```

```python
def marbles_game(players):
    # assert right number of players
    assert (players > 1 and players < 5), 'Invalid number of players'
    
    best = float('Inf')
    player = ''
    
    # simulate games
    for i in range(players):
        points = simulate()
        turns = [i for i in range(len(points))]
        
        # check for the best results
        if len(points) < best:
            best = len(points)
            player = str(i)
            
        # plot
        plt.plot(turns, points, label=str(i))
    
    plt.ylabel('Points')
    plt.xlabel('Turns')
    plt.legend()
    plt.title('Marbles Game')
    plt.show()
    
    return player
        
    
    
    
```
